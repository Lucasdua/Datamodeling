```python
import requests
import pandas as pd

# URL da API do IBGE
url = "https://servicodados.ibge.gov.br/api/v1/localidades/municipios"

# Solicitação para obter os dados
response = requests.get(url)

# Verifica se a solicitação foi bem-sucedida
if response.status_code == 200:
    data = response.json()
    # Converte para DataFrame
    df = pd.json_normalize(data)
    # Salva os dados
    df.to_csv('municipios_ibge.csv', index=False)
    print("Download concluído!")
else:
    print("Erro ao acessar a API")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
df = pd.read_csv('municipios_ibge.csv')

# Visualizar as primeiras linhas
print(df.head())

# Contar os estados
print(df['microrregiao.mesorregiao.UF.nome'].value_counts())

# Gráfico de distribuição de municípios por estado
df['microrregiao.mesorregiao.UF.nome'].value_counts().plot(kind='bar', figsize=(12, 6))
plt.title('Quantidade de Municípios por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Municípios')
plt.show()
```
